## Question 4 (10 points):

Use matplotlib to plot a scatterplot between the Ozone and Wind values to study the relationship between the two by specifying the appropriate DataFrame columns to matplotlib. 

Give the axes appropriate captions and the graphs appropriate titles. 

Place useful comments in your code so that you would remember your programming logic if you were to revisit this code later.

```python
#Solution

#importing Mathplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt

#defining variables and preparing data
Day=ozone_data["Day"]
Ozone=ozone_data["Ozone"]
Wind=ozone_data["Wind"]
x=['Day']
y=['Ozone']
z= ['Wind']

#Plot 1 shows relation between ozone and wind.  Graph shows a negative correlation.
dataFrame = pd.DataFrame(data=ozone_data, columns=['Ozone','Wind']) #defining dataframe
dataFrame.plot.scatter(y,z) #plotting dataframe
plt.title("Ozone Concentration vs. Wind") #chaning title
plt.xlabel("Ozone Concentration") #changing x axis label
plt.ylabel("Wind Speed") #changing y axis label
plt.show #showing as output

#Plot 2 shows Day vs. Ozone values, which I think would be interesting to see.
dataFrame = pd.DataFrame(data=ozone_data, columns=['Day','Ozone'])
dataFrame.plot.scatter(x,y)
plt.title("Day vs. Ozone Concentration")
plt.ylabel("Ozone Concentration")
plt.show

#Plot 3 shows Day vs. Wind values, which I also thought could be interesting to see
dataFrame = pd.DataFrame(data=ozone_data, columns=['Day','Wind'])
dataFrame.plot.scatter(x,z)
plt.title("Day vs. Wind")
plt.ylabel("Wind Speed")
plt.show
```
